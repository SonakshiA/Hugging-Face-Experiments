**Pipeline Usage**

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis') #Uses a default model - distilbert-base-uncased-finetuned-sst-2-english
```

```python
classifier('Thanks a lot for watching the video. Really appreciate it!')
```

```python
results = classifier(["Thanks a lot guys","This video is not cool","I hate my dog","This video is so sick"])
for result in results:
  print(f"Label: {result['label']}, Score: {round(result['score'],4)}")
```
